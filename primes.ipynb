```python
# "a" positive integer

for a in range(1, 100):
    if(a == 1):
        b = 1
    else:
        b = a - 1
    c = 1
    c1 = complex(a, b)
    c2 = complex(a, -c)
    p = c1 * c2
    print("Re[({} + {}i)({} - {}i)] = {}".format(a, b, a, c, p.real))
    c1 = complex(a, b)
    c2 = complex(a, c)
    p = c1 * c2
    print("Re[({} + {}i)({} + {}i)] = {}".format(a, b, a, c, p.real))
```

```python
# "a" is odd positive integer

for a in range(1,100, 2):
    for b in range(1, 100):
        c = b - 1        
        if(b+c == a):
            c1 = complex(a, b)
            c2 = complex(a, -c)
            p = c1 * c2
            print("Re[({} + {}i)({} - {}i)] = {}".format(a, b, a, c, p.real))
            c1 = complex(a, b)
            c2 = complex(a, c)
            p = c1 * c2
            print("Re[({} + {}i)({} + {}i)] = {}".format(a, b, a, c, p.real))
```

```python
#odd and even

for a in range(1,100):
    for b in range(1, 100):
        for c in range(1, 100):
            #c = b - 1        
            if(b+c == a):
                c1 = complex(a, b)
                c2 = complex(a, -c)
                p = c1 * c2
                print("({} + {})({} - {}) = {}".format(a, b, a, c, p.real))
                c1 = complex(a, b)
                c2 = complex(a, c)
                p = c1 * c2
                print("({} + {})({} + {}) = {}".format(a, b, a, c, p.real))
```
